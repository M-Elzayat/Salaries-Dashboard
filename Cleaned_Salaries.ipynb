Preprocessing

1- drop all irrelvant columns

```python
df.drop(columns=['Notes','Agency','Status'],inplace=True)
```

```python
df.columns
```

2- Drop Duplicates

```python
df.drop_duplicates(inplace=True)
```

```python
df.duplicated().sum()
```

Fixed Data types

```python
df['BasePay']=pd.to_numeric(df['BasePay'],errors='coerce')
df['OvertimePay']=pd.to_numeric(df['OvertimePay'],errors='coerce')
df['OtherPay']=pd.to_numeric(df['OtherPay'],errors='coerce')
df['Benefits']=pd.to_numeric(df['Benefits'],errors='coerce')
```

```python
df.dtypes
```

```python
df.describe()
```

Handling Missing Value

```python
#df['BasePay'].fillna(df['BasePay'].median(),inplace=True) to fill all NaN values with median
df['BasePay'].fillna(0,inplace=True)
```

```python
df['Benefits'].fillna(0,inplace=True)
```

```python
df.isnull().sum() # to check again if all null values are removed
```

```python
df.dropna(subset=['OvertimePay','OtherPay'],inplace=True) # delete row nan from column
```
